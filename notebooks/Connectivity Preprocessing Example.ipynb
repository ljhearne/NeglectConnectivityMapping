```python
import pandas as pd

# load participant information

file = '/home/lukeh/projects/NeglectNetworks/data/derivatives/participant_dataframe.csv'

# participant information spreadsheet
df = pd.read_csv(file)
df.head()
```

## Plot a single participant

```python
# plot a random subject
from nilearn import plotting
participant_idx = 0
img = df.raw_lesion_file[participant_idx]
plotting.plot_stat_map(img, cmap='viridis', title='raw lesion')
plotting.show()

img = df.resampled_lesion_file[participant_idx]
plotting.plot_stat_map(img, cmap='viridis', title='resampled and flipped lesion')
plotting.show()

img = df.connectivity_lesion_normalised_file[participant_idx]
plotting.plot_stat_map(img, cmap='viridis', title='connectivity lesion normalised')
plotting.show()

img = df.connectivity_lesion_normalised_binarized80_file[participant_idx]
plotting.plot_stat_map(img, cmap='viridis', title='connectivity lesion binarized')
plotting.show()
```

## Plot group level coverage maps

```python
from nilearn.image import mean_img
img_mean = mean_img(list(df.resampled_lesion_file))
plotting.plot_stat_map(img_mean, cmap='viridis', vmax=0.20)
plotting.show()
```

```python
img_mean = mean_img(list(df.connectivity_lesion_normalised_binarized50_file))
plotting.plot_stat_map(img_mean, cmap='viridis', vmax=0.20)
plotting.show()
```
