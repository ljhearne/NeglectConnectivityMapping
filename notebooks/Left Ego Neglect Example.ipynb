# Left egocentric neglect sanity check

```python
import sys
import nibabel as nib
sys.path.append("../code/")
res_dir = '/home/lukeh/projects/NeglectNetworks/data/derivatives/results/LEgoBin/'
```

## Traditional univariate VLSM
- minimum subjects per voxel = 10
- Bonferroni corrected for multiple comparisons
- Lesion size regressed from behaviour

```python
from nilearn import plotting
from nilearn.image import new_img_like

analysis = 'univariate_lesion'
# raw statistical map
img = res_dir+analysis+'/mask_img.nii.gz'

plotting.plot_stat_map(img, cmap='viridis', draw_cross=False, colorbar=True)
plotting.show()

img = res_dir+analysis+'/zmap_img.nii.gz'
plotting.plot_stat_map(img, vmax=10, cmap='viridis', draw_cross=False, colorbar=False)
plotting.show()

p_mask = nib.load(res_dir+analysis+'/pval_img.nii.gz').get_fdata()
img = nib.load(res_dir+analysis+'/zmap_img.nii.gz').get_fdata()
img[p_mask > 0.01] = 0
img = new_img_like(res_dir+analysis+'/pval_img.nii.gz', img)
plotting.plot_stat_map(img, cmap='viridis', draw_cross=False, colorbar=False)
plotting.show()
```
